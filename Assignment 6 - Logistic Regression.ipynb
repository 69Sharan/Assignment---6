```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
data = pd.read_csv('bank-full.csv', sep=';')
data
```

```python
# Performing One-Hot Encoding for Columns containg Categorical Data

data1 = pd.get_dummies(data,columns=['job','marital','education','contact','poutcome'])
data1
```

```python
pd.set_option("display.max.columns", None)
data1
```

```python
data1.info()
```

### Manually Encoding the following Columns into 0,1 especially target variable 'y'

```python
# Manually Encoding the variables/columns 

data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

```python
data1['month'].value_counts()
```

```python
order = {'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
order
```

```python
data1 = data1.replace(order)
data1
```

```python
data1.info()
```

### Model Building

```python
features = pd.concat([data1.iloc[:,0:11],data1.iloc[:,12:]],axis=1)
features
```

```python
target = data1.iloc[:,11]
target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,target,train_size=0.7,random_state=10)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model = LogisticRegression()
```

```python
import warnings
warnings.filterwarnings('ignore')

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_pred_df = pd.DataFrame({'actual_y':y_test,'pred_y':y_pred})
y_pred_df
```

```python
model.intercept_
```

```python
model.coef_
```

### Testing Model Accuracy

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
confusion_matrix = confusion_matrix(y_test,y_pred)
confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
model.predict_proba(features)[:,1] 
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# Plotting ROC Curve and finding AUC Value

fpr,tpr,thresholds = roc_curve(target,model.predict_proba(features)[:,1])
plt.plot(fpr,tpr,color='red')
auc = roc_auc_score(y_test,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
